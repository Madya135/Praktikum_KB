```python
#Baca berkas CSV
import pandas as pd
rating = pd.read_csv("MALratings.csv")
rating
```

```python
#Mencari Jumlah Record, nama atribut, total objek, total angka, dan total null
jumlahrc = len(rating)
atribut = rating.columns
objek = len(rating.select_dtypes(include=['object']).columns)
float = len(rating.select_dtypes(include=['float64']).columns)
angka = len(rating.select_dtypes(include=['int64']).columns)
null  = len(rating.columns[rating.isna().sum() > 0])

#Dictionary dataset rating
dictionary = {
    "Jumlah Record" : jumlahrc,
    "Nama Atribut" : atribut,
    "Jumlah Atribut Objek" : objek,
    "Jumlah Atribut Angka" : angka,
    "Jumlah Atribut Float" : float,
    "Jumlah Atribut Null" : null,
}
```

```python
#Menampilkan nilai rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.
rating.describe()
```

```python
#Menampilkan isi Dictionary
pd.Series(dictionary)
```

```python
#Mengubah 1 nilai atribut angka menjadi desimal 3 angka dibelakang koma
rating_copy = rating.copy()
rating_copy["Popularity"] /= 1000
rating_copy.head(10)
```

```python
#Menampilkan nilai Korelasi
rating.select_dtypes(include=['int64','float64']).corr()
```

```python
#Menyimpan record dataset yang hanya bertipe angka ke file .CSV baru
rating.select_dtypes(exclude = ["object"]).to_csv("MALratings-number.csv")
```
