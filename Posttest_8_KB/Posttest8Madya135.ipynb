```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
```

```python
# mengatur seed 
np.random.seed(7)
```

```python
# Loading pima indians dataset
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# https://www.kaggle.com/datasets/kumargh/pimaindiansdiabetescsv?resource=download
```

```python
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# mengecek dimensi data X
X.shape
```

```python
# Membuat model
model = Sequential()
model.add(Dense(16, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Ringkasan parameter model
model.summary()
```

```python
# Compiling model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fitting model
history = model.fit(X, Y, validation_split=0.33, epochs=150, batch_size=10)
```

```python
# Grafik Accuracy vs Validation
grafik = range(len(history.history['accuracy']))
plt.plot(grafik, history.history['accuracy'],label='Training')
plt.plot(grafik, history.history['val_accuracy'],label='Validation')
plt.legend()
plt.show()
```
