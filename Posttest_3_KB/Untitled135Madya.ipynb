membuat Grafik Heatmap

```python
sns.heatmap(data=rtp.corr(), cmap="Blues")
plt.show()
```

Membuat Grafik Plot dan Garis

```python
plt.figure(figsize=(7,3))
plt.plot(rtp.head()['Popularity'], rtp.head()['Score'], label = 'Anime Popularity Score', linestyle='--')
plt.legend()
plt.title('Anime Score Popularity')
plt.xlabel('Popularity',fontsize=12)
plt.ylabel('Score Rank',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

Membuat Grafik Scatter Plot

```python
plt.scatter(rtp.head()['Score'], rtp.head()['Popularity'], label='Anime Score Popularity')
plt.legend()
plt.title('Anime Score Popularity')
plt.xlabel('Score Rank',fontsize=12)
plt.ylabel('Popularity',fontsize=12)
```

Membuat Grafik Pie

```python
rt = pd.read_csv("MALratings.csv")

plt.figure(figsize=(7,3))
plt.pie(rt.head(2)['Rank'].value_counts(), labels=['0 - 5.0','5.1  9.9'], startangle=90, autopct= '%.1f%%', pctdistance=0.5, explode=(0.03,0))
plt.title('Anime Score Rank')
plt.legend(loc='lower right')
plt.show()
```
